- Otimized for mathematical and statistical operations.

## Ndarray - The heart of Numpy

An Ndarray (N-dimensional array) is the fundamental data structure in Numpy

it is homogeneous collection of data stored in a grid format

### Creat Array

```python
# 1D array
arr1 = np.array([1,2,3,4])
print("1D array",arr1)
```

```python
# 2D Array
arr2 = np.array([[1,2,3],[4,5,6]])
print("2D array:\n",arr2)
```

```python
#Array if zeros, one, range
zero = np.zeros((2,3))
print(zero)
```

```python
#Array if zeros, one, range
one = np.ones((3,3))
print(one)
```

```python
#Range array
range_array = np.arange(0,10,2)
print(range_array)
```

## Array operation

```python
arr_a = np.array([1,2,3])
arr_b = np.array([10,20,30])

print("Addtion of Array",arr_a + arr_b)
print("Multiplication of Array",arr_a * arr_b)
print("Division of Array",arr_b / 2)
print("squared of Array",arr_a ** 2)
```

```python
# Broad castng
matrix = np.array([[1,2],[3,4]])
print("original matrix\n",matrix)
print()
print("Add scalar 5:\n",matrix + 5)
```

## Indexing, slicing Reshaping

```python
arr = np.array([[10,20,30],[40,50,60]])
print("original array:\n",arr)
print()

# indexing
print("Element [0,1]",arr[0,1])
```

```python
#slicing
print("First row",arr[1:])
```

```python
print("First row",arr[:,-1])
```

```python
print(arr)
```

```python
# Reshaping
reshape = np.arange(12).reshape(3,4)
print(reshape)
```

```python
## Useful funtion of Numpy
arr = np.array([4,7,2,8,3])
print("Mean:",np.mean(arr))
```

```python
print("Standard Deviation:",np.std(arr))
```

```python
print("Standard Deviation:",np.sort(arr))
```

```python
#linespace
lin = np.linspace(0,1,5)
print(lin)
```

```python
# Random
rand = np.random.randint(1,10,(2,3))
print(rand)
```

## Quiz

```python
# can you create matrix with values from 1 to 25
```

```python
matrixQz = np.arange(1,26).reshape(5,5)
print(matrixQz)
```

```python
# get the middle matrix of dimesion 3 X 3
matrixQz = matrixQz[1:4,1:4]
print(matrixQz)
```

```python
# sum of each row
row_sum = np.sum(matrixQz,axis=1)
print("Row wise sum",row_sum)
```

```python
# sum of each column
column_sum = np.sum(matrixQz,axis=0)
print("column wise sum",column_sum)
```
